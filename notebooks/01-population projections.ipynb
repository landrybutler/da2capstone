# Import, clean and review population projection data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Setup filename variables
population_csv = '../data/raw/population_projections_by_county.csv'

# import files to dataframes
population = pd.read_csv(population_csv) 

population
```

```python
nashville_population = population.loc[population.County == 'Davidson']
nashville_population
```

```python
# drop unneded columns
nashville_population = nashville_population.drop(['County FIPS', 'Unnamed: 55', 'Unnamed: 56'], axis=1)

nashville_population
```

```python
# reformat using melt() for plotting in Tableau
# pd.melt(df, id_vars =['Name'], value_vars =['Course', 'Age']) 

nashville_tidy = nashville_population.melt(id_vars =['County'], var_name='year', value_name='population') 
nashville_tidy
```

```python
# save new dataframes
nashville_population.to_csv('../data/clean/nashville_population.csv')
nashville_tidy.to_csv('../data/clean/nashville_tidy_population.csv')
```
