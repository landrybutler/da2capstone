# Prep Files for Tableau

```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# import and set up dataframes
outpatient_clean = pd.read_csv('../data/outpatient_clean.csv', low_memory=False)
providers_clean = pd.read_csv('../data/providers_clean.csv', low_memory=False)
cbsa_to_name = pd.read_csv('../data/core-based-statistical-areas-cbsas-and-combined-statistical-areas-csas.csv', engine='python')
cbsa_excel = '../data/ZIP_CBSA_032020.xlsx'
cbsa = pd.read_excel(cbsa_excel) 

outpatient_clean.head()
```

# Exclude Out-of-State Providers

```python
tennessee_outpatient_clean = outpatient_clean[outpatient_clean ['provider_state'] == 'TN'] 
tennessee_providers_clean = providers_clean[providers_clean ['nppes_provider_state'] == 'TN'] 

tennessee_outpatient_clean.head()
```

```python
tennessee_outpatient_clean.head()
```

# Join Zip and CBSA Files to Get CBSA Names

```python
cbsa_to_name.info()
```

```python
# clean up column names
cbsa.columns = cbsa.columns.str.replace(' ', '_').str.lower()
cbsa_to_name.columns = cbsa_to_name.columns.str.replace(' ', '_').str.lower()
cbsa_to_name['cbsa_code'] = pd.to_numeric(cbsa_to_name['cbsa_code'], errors='coerce')
cbsa_to_name.info()
```

```python
# convert zip code to numeric fields

tennessee_providers_clean.nppes_provider_zip.astype('int64')
tennessee_outpatient_clean.provider_zip_code.astype('int64')

# these lines worked at forst, but not anymore
# tennessee_providers_clean['nppes_provider_zip'].astype('int64')
# tennessee_outpatient_clean['provider_zip_code'].astype('int64')

print(tennessee_providers_clean.info())
print(tennessee_outpatient_clean.info())
```

```python
# join files
tennessee_providers_clean = pd.merge(left=tennessee_providers_clean, right=cbsa, 
                                     left_on='nppes_provider_zip', right_on='zip', how='outer')

tennessee_outpatient_clean = pd.merge(left=tennessee_outpatient_clean, right=cbsa, 
                                      left_on='provider_zip_code', right_on='zip', how='outer')

print(tennessee_providers_clean.info())
print(tennessee_outpatient_clean.info())
```

```python
# Join with cbsa_to_name to add CBSA names

tennessee_providers_clean = pd.merge(left=tennessee_providers_clean, right=cbsa_to_name, 
                                     left_on='cbsa', right_on='cbsa_code', how='inner')

tennessee_outpatient_clean = pd.merge(left=tennessee_outpatient_clean, right=cbsa_to_name, 
                                      left_on='cbsa', right_on='cbsa_code', how='inner')

print(tennessee_providers_clean.info())
print(tennessee_outpatient_clean.info())
```

# Drop Un-needed Columns

```python
# Drop all but listed tables for outpatient
tennessee_outpatient_tableau = tennessee_outpatient_clean[[
    'provider_name',
    'apc',
    'apc_description',
    'beneficiaries',
    'comprehensive_apc_services',
    'average_estimated_total_submitted_charges',
    'average_medicare_allowed_amount',
    'average_medicare_payment_amount',
    'average_medicare_outlier_amount',
    'zip',
    'cbsa',
    'cbsa_title']]
tennessee_outpatient_tableau = tennessee_outpatient_tableau.dropna()
tennessee_outpatient_tableau
```

```python
# Drop all but listed tables for proviers
tennessee_providers_tableau = tennessee_providers_clean[[
    'nppes_provider_last_org_name',
    'provider_type',
    'place_of_service',
    'hcpcs_code',
    'hcpcs_description',
    'line_srvc_cnt',
    'bene_unique_cnt',
    'average_medicare_allowed_amt',
    'average_submitted_chrg_amt',
    'average_medicare_payment_amt',
    'average_medicare_standard_amt',
    'zip',
    'cbsa',
    'cbsa_title']]
tennessee_providers_tableau = tennessee_providers_tableau.dropna()
tennessee_providers_tableau.head(50)
```

# Join Outpatient and Providers

```python
# Rename Columns to Simplify Join
new_outpatient_column_names = [
    'provider_name',
    'apc',
    'apc_description',
    'beneficiaries',
    'comprehensive_apc_services',
    'avg_est_total_submitted_amt',
    'avg_medicare_allowed_amt',
    'avg_medicare_payment_amt',
    'avg_medicare_outlier_amt',
    'zip',
    'cbsa',
    'cbsa_name']

new_provider_column_names = [
    'provider_name',
    'provider_type',
    'place_of_service',
    'hcpcs_code',
    'hcpcs_description',
    'line_srvc_cnt',
    'bene_unique_cnt',
    'avg_medicare_allowed_amt',
    'avg_submitted_chrg_amt',
    'avg_medicare_payment_amt',
    'avg_medicare_standard_amt',
    'zip',
    'cbsa',
    'cbsa_name']

tennessee_outpatient_tableau.columns = new_outpatient_column_names
tennessee_providers_tableau.columns = new_provider_column_names


print(tennessee_providers_tableau.info())
print(tennessee_outpatient_tableau.info())
```

```python
# Concatenate outpatient and providers
tennessee_tableau = tennessee_providers_tableau.append([tennessee_outpatient_tableau], sort=False)

print(tennessee_tableau.info())
```

# Export for Tableau

```python
# save files
tennessee_outpatient_tableau.to_csv('../data/tennessee_outpatient_tableau.csv')
tennessee_providers_tableau.to_csv('../data/tennessee_providers_tableau.csv')
tennessee_tableau.to_csv('../data/tennessee_tableau.csv')
```
